## Script updates a CSV attached to a ArcGIS Online item

with the lastest three titles and URLs to "maps we love"

```python
import datetime
import requests
from arcgis.gis import GIS
gis=GIS('https://xxxxxxxx.maps.arcgis.com','','')

try:
    # get data
    esri=requests.get("https://www.esri.com/en-us/maps-we-love/gallery")
    if esri.status_code != 200:
        payload = "null"
    mapswelove = esri.text
    start1 = mapswelove.find('a href="https://www.esri.com/en-us/maps-we-love/')
    start1b = mapswelove.find('</p>', start1)
    chunk1=mapswelove[start1-1:start1b]
    #print(chunk1)
    # get URL
    url1=chunk1.find('href=')
    url1b=chunk1.find('class=', url1)
    p1=chunk1[url1+6:url1b-2]
    # get image
    url1=chunk1.find('src=') +1
    url1b=chunk1.find('.jpg', url1)
    p2 = 'http://www.esri.com/' + chunk1[url1+4:url1b+4]
    # get linkname
    url1=chunk1.find(':14}]')+7
    p3 = chunk1[url1:]
    payload = '<a href="' + p1 + '" target="new">' + '<img width=300 src="' + p2 + '"><br><br>' + p3 + '</a><br><br>'

    # map 2 --------------------------
    start1 = mapswelove.find('a href="https://www.esri.com/en-us/maps-we-love/', start1b)
    start1b = mapswelove.find('</p>', start1)
    chunk1=mapswelove[start1-1:start1b]
    #print(chunk1)
    # get URL
    url1=chunk1.find('href=')
    url1b=chunk1.find('class=', url1)
    p1=chunk1[url1+6:url1b-2]
    # get image
    url1=chunk1.find('src=') +1
    url1b=chunk1.find('.jpg', url1)
    p2 = 'http://www.esri.com/' + chunk1[url1+4:url1b+4]
    # get linkname
    url1=chunk1.find(':14}]')+7
    p3 = chunk1[url1:]
    payload = payload + '<a href="' + p1 + '" target="new">' + '<img width=300 src="' + p2 + '"><br><br>' + p3 + '</a><br><br>'

    #write to temp file ---------------------------------------
    f = open("home/mapswelove.txt", "w")
    f.write(payload)
    f.close()
except:
    # this writes some maps we love to the file, should the above fail
    f = open("home/mapswelove.txt", "w")
    f.write('<a href="https://www.esri.com/en-us/maps-we-love/gallery/human-reach" target="new"><img width=300 src="http://www.esri.com//content/dam/esrisites/en-us/maps-we-love/gallery/73-the-human-reach.jpg"><br><br>The Human Reach</a><br><br><a href="https://www.esri.com/en-us/maps-we-love/gallery/one-city-marathon" target="new"><img width=300 src="http://www.esri.com//content/dam/esrisites/en-us/maps-we-love/72-one-city-marathon/72-one-city-marathon-tip-3.jpg"><br><br>Newport News One City Marathon</a><br><br>')
    f.close()

# copy file contents from workspace to a plain text file attached to an AGO item (not a published service)

# Item Added From Toolbar
# Title: mapswelove | Type: CSV | Owner: tbakerk12
item = gis.content.get("45f7b3b45a2048f6b6cd9d849342f97c")
item.update(item_properties=None, data='/arcgis/home/mapswelove.txt', thumbnail=None, metadata=None)
```
